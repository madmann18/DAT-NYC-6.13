2. Create three new columns:
   1. `rooms_per_household`
   2. `bedrooms_per_room`

```python
plt.hist(np.log( df['total_rooms']), bins = 30);
print('The skew before was', df['total_rooms'].skew(), '\nAfter transforming is', np.log(df['total_rooms'].skew()))
```

```python
#Draw a scatter plot of the houses location
#coloring them based on the median house value

plt.scatter(df.latitude, df.longitude, c = df.median_house_value)
```

```python
#Transform the tail heavy distributions using mathematical operations
#ln..

sns.lmplot('latitude', 'longitude', data = df)
```

```python
#Create three new columns:
#rooms_per_household
df['rooms_per_household'] = df.total_rooms/df.households
#bedrooms_per_room
df['bedrooms_per_room'] = df.total_bedrooms/df.total_rooms
#population_per_household
df['population_per_household'] = df.population/df.households
df.head()
```

```python
#Drop values where na's are missing, assign this DataFrame to dropped_houses.
dropped_houses = df.dropna()
dropped_houses.head()
```

```python
#correlation for all values in df
corr = df.corr()
```
